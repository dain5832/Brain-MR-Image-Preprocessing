# Mask_processing

Load tumor mask. \
For each pixel, assign 1 if the pixel value is larger than 0.5 and 0 other wise.

```python
import os
import matplotlib.pyplot as plt
import nibabel as nib
import numpy as np
```

```python
sourceDir = 'E:/IDHwt_nii_preprocessed/Tumor_masks_resampled'
targetDir = 'E:/IDHwt_nii_preprocessed/Tumor_masks'
```

```python
# import mask file
total = 0

for (path, dirs, files) in os.walk(sourceDir):
#    if total > 2:
#        break
    for file in files:
        print("Processing mask #{} {}".format(total, file))
        total += 1
        filePath = os.path.join(path, file)
        mask = nib.load(filePath)
        mask_data = mask.get_fdata()
        mask_data_processed = np.where(mask_data < 0.5, 0, 1)
#        fig, axes = plt.subplots()
#        plt.imshow(mask_data[:, :, 48], cmap="gray", origin="lower")
#        fig, axes = plt.subplots()
#        plt.imshow(mask_data_processed[:, :, 50], cmap="gray", origin="lower")
#        plt.title(file)

        
        masked_data = nib.Nifti1Image(mask_data_processed, mask.affine, mask.header)
        nib.save(masked_data, os.path.join(targetDir, file))

    
```
