```python
import os
import pandas as pd
import shutil
from glob import glob
```

```python
in_path = '/media/sc-mlsg/Brain2020/Registry_nii_sorted_preprocessed/01_Resampled/'
out_path = '/media/sc-mlsg/Brain2020/Registry_nii_sorted_preprocessed2/01_Resampled/'

csv_path = '/media/sc-mlsg/Brain2020/Registry_nii_sorted_preprocessed/Registry_nii_code.csv'
os.mkdir(out_path)
```

```python
for (path, dirs, files) in os.walk(in_path):
    print(path, dirs, files)
```

```python
# if csv not exists
pt_lst = sorted([int(i) for i in os.listdir(in_path)])
id_lst = ['T00{}'.format(i) for i in range(1,10)] + ['T{}'.format(i) if i >= 100 else 'T0{}'.format(i) for i in range(10,len(pt_lst)+1)]

df = pd.DataFrame()
df['ID'] = id_lst
df[''] = pt_lst
df.to_csv(csv_path, index=False)
```

```python
code_lst = pd.read_csv(csv_path)
code_lst.head()
```

```python
for (path, dirs, files) in os.walk(in_path):
    if not dirs:
        for file in files:
            pt = str(file.split('_')[0])
            ID = code_lst[code_lst[''] == int(pt)]['ID'].to_list()[0]
            print(pt, ID)
            
            newfilename = '{}_{}'.format(ID, '_'.join(file.split('_')[1:]))
            print(newfilename)
            if not os.path.exists(os.path.join(out_path, ID)):
                os.mkdir(os.path.join(out_path, ID))
            if os.path.exists(os.path.join(out_path, ID, newfilename)):
                print('file already exists')
                continue
            shutil.copyfile(os.path.join(path, file), os.path.join(out_path, ID, newfilename))
```
