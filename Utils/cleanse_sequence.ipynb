```python
import os
import pandas as pd
import shutil
```

```python
seq_dict = { 'T2':['T2 AX', 'AX T2', 'T2 AX GD', '3D T2 AX', 't2_tse_ax'],
 
'T1': ['3D AX', 'T1 AX', 'AX T1', 't1_tse_ax', 'T1 PRE 3D AX', '3D T1 TFE AX'],

'FLAIR': ['T2 FLAIR AX(11000_2800_110)',
 'T2 FLAIR',
 'T2 FLAIR NEW',
 'T2 FLAIR AX',
 't2_tse_FLAIR_ax',
 'AX FLAIR'],

'3D_FLAIR': ['3D_Brain_View_FLAIR AX GD', 
                 '3D Brain View FLAIR AX GD', 
                 'V3D Brain View FLAIR AX GD', 
                 '3D_Brain_View_FLAIR AX'],
 
'BB': ['3D T1 BB MSDE AX F GD',
 '3D T1 BB iMSDE AX F GD',
 'V3D T1 BB iMSDE AX F GD',
 't1-space_3D_BB_ax gd',
 't1_dante_space_sag_bb_iso_gd_MPR_Tra',
 't1_space_3D_BB_sag gd_MPR_Tra'],

'WB': ['3D T1 TFE AX F GD',
 'V3D T1 TFE AX F GD',
 '3D T1 TFE AX GD',
 'T1 Post 3D AX F',
 'T1 POST 3D AX F',
 't1_3D AX GD FS 1mm G2',
 't1_3D SAG GD FS_MPR_Tra']}
```

```python
in_path = '/media/sc-mlsg/easystore/KHS_control_nii/'
out_path = '/media/sc-mlsg/easystore/KHS_control_nii_sorted/'

os.makedirs(out_path, exist_ok=True)
```

```python
for (path, dirs, files) in os.walk(in_path):
    print(path, dirs, files)
```

```python
# 1) create csv file w/ original seq name for each pt
seq_lst = []
for (path, dirs, files) in os.walk(in_path):
    if files:
        pt = path.split(os.path.sep)[-1]
        pt_seq_lst = [pt] + ['_'.join(file.split('.')[0].split("_")[1:]) for file in sorted(files)]
        seq_lst.append(pt_seq_lst)
        
seq_df = pd.DataFrame(seq_lst).rename(columns={0:'pt'})
seq_df['pt'] = seq_df['pt'].astype('int')
seq_df = seq_df.sort_values('pt', axis=0)
seq_df.to_csv(os.path.join(in_path,'original_seq.csv'), index=False)
```

```python
# 2) cleanse sequence
for (path, dirs, files) in os.walk(in_path):
    if not files:
        continue
    for file in files:
        pt = path.split(os.path.sep)[-1]
        seq = '_'.join(file.split('_')[1:]).split('.')[0]
#        print(pt, seq)
        for (key, item) in seq_dict.items():
            if seq in item:
                newfilename = '{}_{}.nii.gz'.format(pt, key)
                print(newfilename)
                if not os.path.exists(os.path.join(out_path, pt)):
                    os.mkdir(os.path.join(out_path, pt))
                if os.path.exists(os.path.join(out_path, pt, newfilename)):
                    print('file already exists')
                    continue
                shutil.copyfile(os.path.join(path, file), os.path.join(out_path, pt, newfilename))
```

```python
# 3) create csv file w/ cleansed seq name for each pt
seq_lst = []
for (path, dirs, files) in os.walk(out_path):
    if files:
        pt = path.split(os.path.sep)[-1]
        pt_seq_lst = [pt] + ['_'.join(file.split('.')[0].split("_")[1:]) for file in sorted(files)]
        seq_lst.append(pt_seq_lst)
        
seq_df = pd.DataFrame(seq_lst).rename(columns={0:'pt'})
seq_df['pt'] = seq_df['pt'].astype('int')
seq_df = seq_df.sort_values('pt', axis=0)
seq_df.to_csv(os.path.join(out_path,'final_seq.csv'), index=False)
```
