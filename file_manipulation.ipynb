```python
import pandas as pd

df1 = pd.read_csv('/home/scmlsg/ResearchProjects/check_sequences200524.csv')
df2 = pd.read_csv('/home/scmlsg/ResearchProjects/added.csv')
```

```python
list_1 = df1.iloc[:, 0]
list_2 = df2.iloc[:, 0]

pt_mrn = set(pd.concat([list_1, list_2], axis=0))
```

```python
print(len(list_1), len(list_2), len(pt_mrn))
```

```python
import os

srcDir = '/media/scmlsg/easystore/BrainMRI-RawData/BrainMRI-Severance/SEV_bMRI_nifti'

total = set(sorted(os.listdir(srcDir))[1:])

```

```python
already = set([str(pt) for pt in list(pt_mrn)])
already
```

```python
not_yet = list(total.difference(already))
set_1 = not_yet[:141]
set_2 = not_yet[141:282]
set_3 = not_yet[282:]

print(len(set_1), len(set_2), len(set_3))
```

```python
import shutil
import os



for pt in list(total):
    srcPath = os.path.join(srcDir, pt)
    
    if pt in set_1:
        targetDir = '/media/scmlsg/Chansik/three_sets_GREMR/CA'
        if not os.path.exists(targetDir):
            os.makedirs(targetDir)
            
        targetPath = os.path.join(targetDir, pt)
        if os.path.exists(targetPath):
            continue
        shutil.copytree(srcPath, targetPath)
        
    if pt in set_2:
        targetDir = '/media/scmlsg/Chansik/three_sets_GREMR/YWP'
        if not os.path.exists(targetDir):
            os.makedirs(targetDir)
        targetPath = os.path.join(targetDir, pt)
        if os.path.exists(targetPath):
            continue
        shutil.copytree(srcPath, targetPath)
        
    if pt in set_3:
        targetDir = '/media/scmlsg/Chansik/three_sets_GREMR/SSA'
        if not os.path.exists(targetDir):
            os.makedirs(targetDir)
        targetPath = os.path.join(targetDir, pt)
        if os.path.exists(targetPath):
            continue
        shutil.copytree(srcPath, targetPath)
    
    
    
    
```
